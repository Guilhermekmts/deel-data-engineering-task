# DEEL Spark CDC Exploration Notebook
This notebook runs inside the Spark container and can query both Kafka CDC topics and the final analytics layer in Postgres.

```python
import os
from pyspark.sql import SparkSession
from delta.pip_utils import configure_spark_with_delta_pip

builder = (
    SparkSession.builder
    .appName("deel-notebook")
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
)
spark = configure_spark_with_delta_pip(builder).getOrCreate()
# spark.sparkContext.setLogLevel("WARN")
spark
```

## Query Delta silver table

```python
delta_path = os.getenv("DELTA_ROOT", "/workspace/data/delta") + "/silver_orders"
orders_df = spark.read.format("delta").load(delta_path)
orders_df.count()
```

```python
delta_path = os.getenv("DELTA_ROOT", "/workspace/data/delta") + "/silver_order_items"
orders_items = spark.read.format("delta").load(delta_path)
orders_items.count()
```
